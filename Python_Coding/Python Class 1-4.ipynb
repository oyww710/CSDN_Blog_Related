```python
weight=input('weight?')
height=input('height?')
w=float(weight)
h=float(height)
result=w/h/h
if result<18.5:
    print('less than average')
elif result<=25:
    print('average')
elif result<=30:
    print('above average')
elif result<=40:
    print('much over average')
else:
    print('extreme above average')
```

Break: 立刻跳出循环

```python
for letter in 'laioffer':
    if letter=='i':
        break
    print('letter is:', letter)
```

Continue: 跳过执行当前，执行下一个

```python
for letter in 'laioffer':
    if letter=='i':
        continue
    print('letter is: ',letter)
```

Break and Continue只针对上紧接着的一个while/for loop

```python
for i in range(1,3):
    for letter in 'laioffer':
        if letter=='i':
            break
        print('letter is:', letter)
    print(i)
```

```python
for i in range(1,3):
    for letter in 'laioffer':
        if letter=='i':
            continue
        print('letter is:', letter)
    print(i)
```
